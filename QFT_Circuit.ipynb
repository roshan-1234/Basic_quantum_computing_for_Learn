```python
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, Aer, transpile, assemble, execute
from qiskit.visualization import plot_histogram
import numpy as np
```

```python
# Create a quantum circuit with the given number of qubits
qc = QuantumCircuit(3)
```

```python
qc.h(0)
qc.draw()
```

```python
qc.cp(np.pi/2, 0, 1) # CROT from qubit 1 to qubit 2
qc.draw()
```

```python
qc.cp(np.pi/4, 0, 2) # CROT from qubit 2 to qubit 0
qc.draw()
```

```python
qc.h(1)
qc.cp(np.pi/2, 1, 2) # CROT from qubit 0 to qubit 1
qc.h(2)
qc.draw()
```

```python
def myqft(nqubits):
    # Create a quantum circuit with the given number of qubits
    myqft_circuit=QuantumCircuit(nqubits)
    
    for qubit in range(nqubits):
        # Apply Hadamard gate to all qubits
        myqft_circuit.h(qubit)
        
        for otherqubit in range(qubit+1,nqubits):
            # Apply controlled-phase gates
            myqft_circuit.cp(np.pi/(2**(otherqubit-qubit)),otherqubit,qubit)
    
    
    return myqft_circuit

display(myqft(3).draw())
            
        
```
